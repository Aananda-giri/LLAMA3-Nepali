## Save iriis data to csv file

```python
from datasets import load_dataset
import csv

# Load the dataset in streaming mode
train_dataset = load_dataset("IRIISNEPAL/Nepali-Text-Corpus", split="train", streaming=True)
test_dataset =  load_dataset("IRIISNEPAL/Nepali-Text-Corpus", split="test", streaming=True)
# Output file path
output_csv = "IRIISNEPAL_Nepali_Text_Corpus.csv"

# Write the dataset to a single CSV file
with open(output_csv, mode='w', newline='', encoding='utf-8') as csvfile:
    writer = None
    for dataset in [train_dataset, test_dataset]:
        for i, sample in enumerate(dataset):
            # Initialize CSV writer with headers on the first iteration
            if writer is None:
                writer = csv.DictWriter(csvfile, fieldnames=sample.keys())
                writer.writeheader()

            # Write each row to the CSV
            writer.writerow(sample)

            # Optional: Print progress every 10,000 rows
            if (i + 1) % 50_000 == 0:
                print(f"{i + 1} rows written...")
```
