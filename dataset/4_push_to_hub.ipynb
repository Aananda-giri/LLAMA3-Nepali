## convert csv to partquet. [reference]()

```python
import duckdb
import time

duck_time = time.time()
con = duckdb.connect(database=':memory:')

con.execute("""
 COPY (select * FROM read_csv('tokenized_data.csv', delim='\t', header=True,  AUTO_DETECT=TRUE, quote='', SAMPLE_SIZE=1000000))
 TO 'tokenized_data.parquet'
 (FORMAT PARQUET, CODEC 'SNAPPY');
""")
duck_time = time.time() - duck_time

print(f"Duckdb time: {duck_time:.2f} seconds")
```
